```python
import pandas as pd
import pylab as pl
import seaborn as sn
pl.style.use('bmh')

T=pd.read_csv(r'C:\Users\julia\Documents\python36\datasource\data\criquets.csv')
T.head()

```

```python
T.plot.scatter(x = 'X', y = 'Y')
```

```python
 # Calculate mean of X and Y columns
mx = T['X'].mean()
mx
```

```python
my = T['Y'].mean()
my
```

```python
# Calculate standard deviation of X and Y columns
sx = T['X'].std(ddof=0)
sx
```

```python
sy = T['Y'].std(ddof=0)
sy

```

```python
# Normalize the X and Y columns
T['NX'] = (T['X'] - mx) / sx
T.head()
```

```python
T['NY'] = (T['Y'] - my) / sy

T.head()
```

```python
# Create scatter plot of normalized X and Y columns
T.plot.scatter(x='NX', y='NY')

# Add x and y axis to the plot
pl.axhline(y=0, color='gray')
pl.axvline(x=0, color='gray')

```

```python
# Calculate covariance of normalized X and Y columns
cov = (T['NX'] * T['NY']).mean()
cov
```

```python
# Create scatter plot of normalized X and Y columns
T.plot.scatter(x='NX', y='NY')

# Add x and y axis to the plot
pl.axhline(y=0, color='gray')
pl.axvline(x=0, color='gray')

# Plot line with slope equal to the covariance
pl.plot([-2, 2], [-2 * cov, 2 * cov], color='red')
```

```python
def R(m):
    carre = (T['NY'] - m * T['NX'])**2
    return carre.sum()
```

```python
M = [k/1000 for k in range(2001)]
LR = [R(m) for m in M]
pl.plot(M, LR, color='black')
```
